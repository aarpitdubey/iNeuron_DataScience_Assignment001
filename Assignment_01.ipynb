<p style='text-size:10px'><b>Ques 2. Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.</b></p>

```python
# Step 1: Creading a list of number 
#from 2000 to 3200 and storing them to nums var

nums = [i for i in range(2000, 3201)] 

#Step 2: using for loop iterate them one by one
for num in nums:
    # Step 3: taking modulus with 7 for checking divisibility
    if num % 7 == 0:
        # Step 4: Checking num should not be the multiple of 5
        if num % 5 != 0:
            # Step 5: Simply printing the num
            print(num, end=', ')
```

<b>Steps: </b>
<b>Method 1: </b>
> 1. Creading a list of number from 2000 to 3200 and storing them to nums var.
> 2. Take a range b/w 2000 t0 3200 (3200+1 including) iterate using for loop
> 3. Taking modulus with 7 for checking divisibility
> 4. Checking num should not be the multiple of 5 but checking num%5 is not equal to 0
> 5. Print the number 

```python
# Step 1: Make an empty list
lst = list()

# Step 2: Take a range b/w 2000 t0 3200 (3200+1 including) iterate using for loop
for num in range(2000, 3201):
    # Step 3: putting condition 1 and 2
    if num%7==0 and num%5!=0:
        # Step 4: append the empty list adding the number 
        lst.append(str(num))
#  Step 5: print the nun and join commas
print(', '.join(lst))
```

<p><b>Ques 2. Create the below pattern using nested for loop in Python. </b></p>



/

```python
n = 5
for i in range (n):
    for j in range(i+1):
        print("*", end=' ')
    print()

for p in range (n, 0, -1):
    for q in range(p-1):
        print("*", end=' ')
    print()
```

> 3. From 2nd step we can find the half part of pattern now for same we have to decrement by 1
> 4. range same but opposite side means step size is -1 and starting from 0th element

<p><b>Write a Python program to reverse a word after accepting the input from the user. 
 
Sample Output: 
 
Input word: AcadGild 
 
Output: dilGdacA</b></p>

```python
def split_(var):
    return [var for var in input_]

def swap_(list, loc1, loc2): 
    list[loc1], list[loc2] = list[loc2], list[loc1] 
    return list

input_     = input("\ninput word: ")
input_     = split_(input_)
input_     = input_[::-1]
print("\nReverse: ", input_)
loc1, loc2 = 1, 3
input__    = swap_(input_, loc1, loc2-1)

print('\nOutput: ', end='')
for input_ in input__:   
    print(input_, end='')
```

<b>Steps: </b>

> 1. Creating two functions first for splitting the char and second one for swapping the char.
> 2. Taking the input
> 3. Using Slicing [: : -1] reversing that input
> 4. Giving location of char for swaping
> 5. Using a for loop printing the values.
